```python
from sklearn.preprocessing import StandardScaler
```

```python
features = ['popularity', 'duration_ms', 'explicit', 'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 
            'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature']

features_df = songs_df[features]
scaler = StandardScaler()

features_scaled_df = pd.DataFrame(scaler.fit_transform(features_df), columns=features)
```

```python
from re import compile as rcompile

rex = rcompile('[^a-zA-Z 0-9]')

tokenize = lambda x: rex.sub('', x.lower().replace(',', ''))

songs_df['genres_tokens'] = songs_df.genres.apply(tokenize)
```

```python
print(songs_df.shape)
songs_df.head()
```

```python
genres_data = songs_df.genres_tokens.to_list()
display(genres_data[:10])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Instantiate vectorizer object

genres_tfidf = TfidfVectorizer(stop_words='english',
                               ngram_range=(1,2),
                               min_df=3,
                               )

# Create a vocabulary and get word counts per document

genres_dtm = genres_tfidf.fit_transform(genres_data)

genres = genres_tfidf.get_feature_names()
#display(len(features), features[:50])

# Get feature names to use as dataframe column headers
genres_dtm = pd.DataFrame(genres_dtm.todense(), columns=genres)
print(genres_dtm.shape)
display(genres_dtm.head())
```

```python
# Use sklearn transformations to merge the features
# I used this article to do this
# https://towardsdatascience.com/how-to-combine-textual-and-numerical-features-for-machine-learning-in-python-dc1526ca94d9
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.preprocessing import FunctionTransformer

numeric_features = [
    'popularity', 'duration_ms', 'explicit', 'release_date',
    'danceability', 'energy', 'key', 'loudness',
    'mode', 'speechiness', 'acousticness', 'instrumentalness',
    'liveness', 'valence', 'tempo', 'time_signature'
]

def get_numeric(df):
    df.release_date = df.release_date.apply(
        lambda date: int(date[:4])
    )
    return df[numeric_features]

def get_text(df):
    return df.genres_tokens

# Create transformers
numeric_transformer = FunctionTransformer(get_numeric)
text_transformer = FunctionTransformer(get_text)

# Create pipeline
feature_pipeline = Pipeline([
    ('features', FeatureUnion([
        ('numeric_features', Pipeline([
            ('numeric_transformer', numeric_transformer),
            ('scaler', StandardScaler())
        ])),
        ('text_features', Pipeline([
            ('text_transformer', text_transformer),
            ('vect', TfidfVectorizer())
        ]))
    ]))
    # Add model here, or connect this pipeline to another pipeline
])

X = feature_pipeline.fit_transform(songs_df)
X
```
